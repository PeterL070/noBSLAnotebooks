# Mean and Variance using vector projections

SymPy calculations of the twitter thread by @MrHonner: https://twitter.com/MrHonner/status/1503152419101941762

```python
from sympy import *
import statistics
Vector = Matrix
```

Consider some arbitrary vector of data.



```python
# data vector
values = [4.0, 7.0, -3.0, 6.0, 1.0]
v = Vector(values)
v
```

```python
# ones vector of the same dimension
a = Vector([1]*len(v))
a
```

```python
mean = v.dot(a)/a.dot(a)
mean
```

```python
# cf.
statistics.mean(values)
```

The remainder of the vector is



```python
dev = v - mean*a
dev
```

```python
dev.norm()**2 / 5
```

```python
# cf.
statistics.pvariance(values)
```

The deviation vector is orthogonal to the mean:



```python
dev.dot( mean*a )
```
